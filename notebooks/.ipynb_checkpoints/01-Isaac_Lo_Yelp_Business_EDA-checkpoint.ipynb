# Load Business Data and Clean It Up

Data location: https://www.yelp.com/dataset

```python
with open('../data/yelp_dataset/yelp_academic_dataset_business.json', 'r') as f:
    data_business = f.readlines()
```

```python
data = [json.loads(business) for business in data_business]
data[0]
```

To find APPI businesses I will want to look at "categories" however this is a string of multiple categories separated by commas. I'm going to turn this into a list.

```python
for bus in data:
    if bus['categories']:
        bus['categories'] = bus['categories'].split(', ')
```

```python
data[5678]
```

I'm going to look at all unique categories

```python
cats = set()
for bus in data:
    categories = bus.get('categories')
    if categories:
        for cat in categories:
            cats.add(cat)
```

```python
len(cats)
```

```python
cats = list(cats)
cats[cats.index('Oncologist'):]
```

```python
# List of categories that are most likely AAPI
aapi_cats = [
    'Chinese', 'Traditional Chinese Medicine', 'Laotian', 'Szechuan', 'Fuzhou', 'Hot Pot', 'Tonkatsu', 'Qi Gong', 'Shanghainese', 'Oriental',
    'Cantonese', 'Pan Asian', 'Sushi Bars', 'Korean', 'Japanese', 'Hainan', 'Taiwanese', 'Asian Fusion', 'Burmese', 'Tai Chi', 'Cambodian', 'Vietnamese',
    'Karate', 'Chinese Martial Arts', 'Dumplings', 'Taekwondo', 'Izakaya', 'Thai', 'Dim Sum', 'Singaporean', 'Noodles', 'Japanese Curry', 'Mongolian',
    'Ramen', 'Indonesian', 'Filipino', 'Hong Kong Style Cafe', 'Donburi', 'Malaysian', 'Karaoke','Teppanyaki', 'Buddhist Temples', 'Muay Thai', 'Poke',
    'Tui Na'
]
```

Create a function to get businesses that contain at least of the categories in `aapi_cats`

```python
def check_aapi(business_cats):
    if any(aapi_cat in business_cats for aapi_cat in aapi_cats):
        return True
    return False

def get_aapi_business(businesses):
    aapi_businesses = []
    for bus in businesses:
        cats = bus.get('categories')
        if cats:
            if check_aapi(cats):
                aapi_businesses.append(bus)

    return aapi_businesses
```

```python
aapi_businesses = get_aapi_business(data)
```

```python
f'{len(aapi_businesses) / len(data):.4f} of businesses are AAPI'
```

```python
f'{len(aapi_businesses)} businesses out of {len(data)} are AAPI'
```

Ill make a dictionary with `business_id` as the key and the business information as the value. I will also make another dictionary with `business_id` as key and a list as the value where the list will contain dictionaries with information regarding a review.

```python
aapi_business_id_map = {buss['business_id']: buss for buss in aapi_businesses}
```

# Load reviews data

```python
with open('../data/yelp_dataset/yelp_academic_dataset_review.json', 'r') as f:
    review_raw = f.readlines()
```

```python
review_data = [json.loads(review) for review in tqdm(review_raw)]
review_data[:10]
```

```python
def map_review_to_business(business_dict, reviews):
    # Iterate through all reviews and if the review's business id is an aapi business
    # add the review to that aapi business's review list.
    business_ids = business_dict.keys()
    for review in tqdm(reviews):
        review_bus_id = review.get('business_id')
        if review_bus_id in business_ids:
            business_dict[review_bus_id].append(review)

    return business_dict
```

```python
from collections import defaultdict
```

```python
business_reviews = {buss: [] for buss in aapi_business_id_map.keys()}
business_reviews = map_review_to_business(business_reviews, review_data)
```

# Basic EDA of reviews

First, ill do a timeseries of average review stars

```python
# First I need to get date without time
df_reviews_['day'] = df_reviews_['date'].dt.to_period('D')
```

```python
df_reviews_.head()
```

```python
avg_stars_by_date = df_reviews_.groupby('day')['stars'].mean()
avg_stars_by_date = avg_stars_by_date.reset_index()
avg_stars_by_date
```

```python
avg_stars_by_date['day'].dtype
```

```python
avg_stars_by_date['day'] =  pd.to_datetime(avg_stars_by_date['day'].astype('string'))
avg_stars_by_date.head()
```

```python
avg_stars_by_date.info()
```

```python
days = avg_stars_by_date['day'].values
stars = avg_stars_by_date['stars'].values
```

```python
fig, axes = plt.subplots(figsize=(15, 8), nrows=3)
axes[0].plot(days[:365], stars[:365])
axes[0].set_title('Avg Rating for AAPI Business by Day in 2019')
axes[0].set_yticks([0,1,2,3,4,5])

axes[1].plot(days[365: 365*2], stars[365: 365*2])
axes[1].set_title('Avg Rating for AAPI Business by Day in 2020')
axes[1].set_yticks([0,1,2,3,4,5])


axes[2].plot(days[365*2: 365*3], stars[365*2: 365*3])
axes[2].set_title('Avg Rating for AAPI Business by Day in 2021')
axes[2].set_yticks([0,1,2,3,4,5])

fig.tight_layout()
plt.savefig('../static/avg_rating_by_day.png', facecolor='white')
plt.show()

```
